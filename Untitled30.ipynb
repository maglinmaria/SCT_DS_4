```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Year_Birth"], bins=20)
plt.title("Distribution of Birth Year")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Income"], bins=30)
plt.show()
```

```python
products = [
'MntWines',
'MntFruits',
'MntMeatProducts',
'MntFishProducts',
'MntSweetProducts',
'MntGoldProds'
]

df[products].sum().plot(kind='bar')
plt.title("Product Sales")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.show()
```

```python
sns.scatterplot(
x="Income",
y="MntWines",
data=df
)
```

```python
campaign = df[
[
'AcceptedCmp1',
'AcceptedCmp2',
'AcceptedCmp3',
'AcceptedCmp4',
'AcceptedCmp5',
'Response'
]
].sum()

campaign.plot(kind='bar')
plt.title("Campaign Responses")
plt.show()
```

```python
print(df.columns.tolist())
```

```python
df["Total_Spent"] = (
    df["MntWines"] +
    df["MntFruits"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)
```

```python
df["Total_Purchases"] = (
    df["NumWebPurchases"] +
    df["NumCatalogPurchases"] +
    df["NumStorePurchases"]
)
```

```python
df.head()
```

```python
products = [
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "MntGoldProds"
]

df[products].sum().plot(kind="bar", figsize=(8,5))
plt.title("Total Spending by Product")
plt.ylabel("Amount")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Income", y="Total_Spent", data=df)
plt.title("Income vs Total Spending")
plt.show()
```

```python
channels = [
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases"
]

df[channels].sum().plot(kind="bar")
plt.title("Purchases by Channel")
plt.ylabel("Total Purchases")
plt.show()
```

```python
campaigns = [
    "AcceptedCmp1",
    "AcceptedCmp2",
    "AcceptedCmp3",
    "AcceptedCmp4",
    "AcceptedCmp5",
    "Response"
]

df[campaigns].sum().plot(kind="bar", color="orange")
plt.title("Campaign Acceptance")
plt.ylabel("Accepted Customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["NumWebVisitsMonth"], bins=15)
plt.title("Website Visits Per Month")
plt.show()
```

```python
channel_data = df[[
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases"
]].sum()

channel_data.plot(kind="bar")
plt.title("Customer Purchase Funnel")
plt.ylabel("Number of Purchases")
plt.show()
```
